### Philadelphia chromosome
A characteristic chromosomal translocation. DNA is "swapped" between chromosomes 9 and 22. This leads to the formation of an oncogene called BCR-ABL which produces the BCR-ABL protein, which is one type of tyrosine kinase. 

The BCR-ABL chimeric gene product has a tyrosine kinase activity several fold higher than it's normal counterpart and correlates with the disease phenotype. This protein affects multiple signaling pathways that directly affect apoptotic potential, cell division rates, and different stages of the cell cycle to achieve unchecked proliferation characteristic (grow and divide out of control) of CML and Ph+ ALL. 

Specifically, JAK / STAT pathway and Ras / MAPK / ERK pathway get affected.

https://www.ncbi.nlm.nih.gov/pmc/articles/PMC1074718/
https://en.wikipedia.org/wiki/Philadelphia_chromosome

## <font color='orange'>Chronic myelogenous/myeloid  leukemia (CML)</font>
A form of leukemia characterized by the increased and unregulated growth of myeloid cells in the bone marrow and the accumulation of these cells in the blood. 

A clonal bone marrow stem cell disorder in which a proliferation 扩增 of mature granulocytes 粒细胞 (neutrophils 中性粒细胞, eosinophils 嗜酸粒细胞 and basophils 嗜碱粒细胞) and their precursors is found. 

A type of myeloproliferative neoplasm associated with Philadelphia chromosome.

CML is largely treated with targeted drugs called TKIs which have led to dramatically improved long-term survival rates since 2001. These drugs have revolutionized treatment of this disease and allow most patients to have a good quality of life when compared to the former chemotherapy drugs. In Western countries, CML accounts for 15–25% of all adult leukemias and 14% of leukemias overall (including the pediatric population, where CML is less common).


#### Symptoms
Fatigue, weakness, night sweats, pale skin color, fever, easy bleeding or bruising, bone pain.

Pain or a sense of "fullness" in the belly; Feeling full after eating even a small amount of food.

Enlarged lymph nodes, spleen (felt as a mass under the left side of the ribcage 胸腔)

#### Lab tests
1. Blood cell counts
- Too many white blood cells with a lot of early (immature) cells called myeloblasts or blasts. Sometimes CML patients have low numbers of red blood cells or blood platelets. 
- Even though these findings may suggest leukemia, this diagnosis usually needs to be confirmed by another blood test or a test of the bone marrow.

2. Bone marrow test
- Cellularity: If your bone marrow sample has more blood-forming cells than expected $\Rightarrow$ hypercellular; if too few of these cells are found $\Rightarrow$ hypocellular.
- Normal bone marrow contains both blood-forming cells and fat cells. In people with CML, the bone marrow is often hypercellular because it's full of leukemia cells.

3. Blood chemistry tests
- Measure the amount of certain chemicals in your blood, but they're not used to diagnose leukemia. They can help find liver or kidney problems caused by the spread of leukemia cells or by the side effects of certain drugs. These tests also help determine if you need to be treated to correct low or high blood levels of certain minerals.

#### Genetic tests
1. Conventional cytogenetics
- Looks at chromosomes (pieces of DNA) under a microscope to find any changes so as to find a Ph chromosome 

2. Fluorescent in situ hybridization (FISH)
- Looks for specific pieces of the BCR-ABL gene on chromosomes.

3. Polymerase chain reaction (PCR)
- A super-sensitive test that can be used to look for the BCR-ABL gene in leukemia cells and measure how much is there even if the amount is small.

#### Imaging tests
1. Computed tomography (CT) scan
- They aren't needed to diagnose CML, but are sometimes used to look for the cause of symptoms or to see if the spleen, liver or lymph nodes are enlarged.

2. Magnetic resonance imaging (MRI)
- very helpful in looking at the brain and spinal cord. 

3. Ultrasound
- look at lymph nodes near the surface of your body or to look for enlarged organs inside your abdomen (belly) such as the kidneys, liver, and spleen. 

#### Possible risk factors
Radiation exposure: Being exposed to high-dose radiation (such as being a survivor of an atomic bomb blast or nuclear reactor accident) increases the risk of getting CML

Age: The risk of getting CML goes up with age

Gender: More common in males than females

https://en.wikipedia.org/wiki/Chronic_myelogenous_leukemia
https://www.ncbi.nlm.nih.gov/pmc/articles/PMC1074718/

#### Phases of CML
1. Chronic phase
- Blasts $\leqslant$ 10% in their blood or bone marrow samples.
- mild symptoms (if any)

2. Accelerated phase
- 15% $\leqslant$ blasts $\leqslant$ 30%, or
- Basophils $\geqslant$ 20% of the blood, or
- Very low platelet counts $\leqslant 100*1000/mm^3$, or
- New chromosome changes in the leukemia cells with the Philadelphia chromosome

3. Blast phase / acute phase / blast crisis
- Blasts $\geqslant$ 20% with large clusters of blasts, or
- Metastasis

#### Adverse prognostic factors
- Accelerated phase or blast phase
- Enlarged spleen
- Areas of bone damage from growth of leukemia
- Increased number of basophils and eosinophils (certain types of granulocytes) in blood samples
- Very high or very low platelet counts
- Age 60 years or older
- Multiple chromosome changes in the CML cells

#### Treatment
1. Targeted Therapies
- TKI drugs that target BCR-ABL are the standard treatment for CML
- In most people, the TKIs don't seem to make the leukemia go away forever, so these drugs need to be taken indefinitely. 

2. Allogeneic Stem cell transplant
- Because collecting the patient’s stem cells can also collect leukemia cells, allogeneic SCT is the only known cure for CML. Still, this type of transplant can cause severe or even life-threatening complications and side effects, and it's often not be a good option in people who are older or have other health problems.

3. Chemo
- Nowadays chemo may be used to treat CML when the TKIs have stopped working. 
- It's also used as part of a stem cell transplant.

4. Radiation Therapy
- Some people have symptoms if swollen internal organs (such as an enlarged spleen) press on other organs. For instance, pressure against the stomach may affect appetite. If these symptoms are not helped by other treatments, radiation to shrink the spleen may be an option.
- Radiation can also be useful in treating pain from bone damage caused by the growth of leukemia cells within the bone marrow.

5. Splenectomy
- If leukemia spreads to the spleen, it can become large enough to compress nearby organs and cause symptoms. If chemotherapy or radiation doesn't help shrink the spleen, it may be removed with surgery.

https://www.cancer.org/cancer/chronic-myeloid-leukemia/treating/targeted-therapies.html

## <font color='orange'>Acute lymphoblastic leukemia (ALL)</font>

A cancer of the lymphoid line of blood cells characterized by the development of large numbers of immature lymphocytes.

As an acute leukemia, ALL progresses rapidly and is typically fatal within weeks or months if left untreated.

#### Symptoms
tired, pale skin color, fever, easy bleeding or bruising, enlarged lymph nodes, or bone pain.

#### Possible risk factors
Genetic: Down syndrome, Li-Fraumeni syndrome, or neurofibromatosis 神经纤维瘤病 type 1. 

Environmental: significant radiation exposure or prior chemotherapy.

Some hypothesize that an abnormal immune response to a common infection may be a trigger.

https://en.wikipedia.org/wiki/Acute_lymphoblastic_leukemia

#### Blood tests
1. Complete blood count (CBC) 
- measures the numbers of red blood cells, white blood cells, and platelets.
2. peripheral blood smear 
- a drop of blood is smeared across a slide  and then looked at under a microscope to see how the cells look 涂片检查. 
- Changes in the numbers and the appearance of the cells often help diagnose leukemia.

Most patients with ALL have too many immature white cells called lymphoblasts (or just blasts) in their blood, and not enough red blood cells or platelets. Lymphoblasts are not normally found in the blood, and they don't function like normal, mature white blood cells.

3. Blood chemistry tests
In patients already known to have ALL, 
- help detect liver or kidney problems caused by spreading leukemia cells or the side effects of certain chemotherapy drugs. 
- help determine if treatment is needed to correct low or high blood levels of certain minerals.

4. Blood coagulation tests
-  make sure the blood is clotting properly.

#### Bone marrow tests
- bone marrow aspiration 刺穿 & bone marrow biopsy 活检
- help diagnose leukemia by determining what percentage of cells in the bone marrow are blasts. 
- They may also be done again later to tell if the leukemia is responding to treatment.

#### Chromosome tests
- a standard part of the work-up for determining ALL pt's outlook and likely response

#### Lumbar puncture (spinal tap)
- check for a spread to brain or spinal cord
- used to put chemotherapy drugs into the CSF to try to prevent/treat this spread

#### Imaging tests
1. X-rays
- if the doctor suspects a lung infection. 
- look for enlarged lymph nodes in the chest.

2. Computed tomography (CT) scan
- show enlarged lymph nodes or organs. 

3. Magnetic resonance imaging (MRI) scan
- looking at the brain and spinal cord if a lumbar puncture finds leukemia cells in the CSF, or if a person is having symptoms that could mean the ALL has spread to the area around the brain.

4. Ultrasound
- used to look at lymph nodes near the surface of the body; enlarged organs inside the abdomen such as the kidneys, liver, and spleen; testicles, if needed.

#### Subtypes of CML
1. B-cell ALL
- B-cell ALL with hypodiploidy (the leukemia cells have fewer than 44 chromosomes)
- B-cell ALL with hyperdiploidy (the leukemia cells have more than 50 chromosomes)
- B-cell ALL with a translocation between chromosomes 9 and 22 (the Philadelphia chromosome, which creates the BCR-ABL1 fusion gene)
- B-cell ALL with a translocation between chromosome 11 and another chromosome
- B-cell ALL with a translocation between chromosomes 12 and 21
- B-cell ALL with a translocation between chromosomes 1 and 19 
- B-cell ALL with a translocation between chromosomes 5 and 14 
- B-cell ALL with amplification (too many copies) of a portion of chromosome 21 (iAMP21)*
- B-cell ALL with translocations involving certain tyrosine kinases or cytokine receptors (also known as “BCR-ABL1–like ALL”)*

2. T-cell ALL 
- Early T-cell precursor lymphoblastic leukemia*
* It's not yet clear if there's enough evidence that it's a unique group

3. Mixed lineage acute leukemias
- A small number of acute leukemias have both lymphocytic and myeloid features. Sometimes the leukemia cells have both myeloid and lymphocytic traits in the same cells. In other cases, a person may have some leukemia cells with myeloid features and others with lymphocytic features.
- Have poorer outlook than standard ALL or AML

#### Prognostic factors
- Age
- Initial white blood cell (WBC) count
- Gene or chromosome abnormalities
- Response to chemotherapy
- Status of ALL during and after treatment
    - Remission, Minimal residual disease (MRD), Active disease

#### Treatment
1. Targeted Therapies - TKIs
- target BCR-ABL. 1/4 ALL pts have Ph+
2. Immunotherapy drugs (IMT)
- Monoclonal antibodies
    - can be designed to attack a specific target, such as a protein on the surface of leukemia cells.
    - used to treat some types of B-cell ALL, typically after chemotherapy has been tried. B cells include some leukemia and lymphoma cells
    - 1. Inotuzumab ozogamicin : This drug is an anti-CD22 antibody linked to a chemotherapy drug. B cells (including some leukemia cells) usually have the CD22 protein on their surface. The antibody acts like a homing signal, bringing the chemo drug to the leukemia cells, where it enters the cells and kills them when they try to divide into new cells. 
    - 2. Blinatumomab : can attach to the CD19 protein, which is found on B cells, AND to CD3, a protein found on T cells. By binding to both of these proteins, this drug brings the cancer cells and immune cells together, which is thought to cause the immune system to attack the cancer cells.
- CAR T-cell therapy
    - Tisagenlecleucel : targets the CD19 protein on certain leukemia cells. It can be used in children and young adults up to age 25 to treat B-cell ALL that has come back after treatment or that is no longer responding to treatment. In this treatment, T cells are removed from the person's blood via leukapheresis and genetically altered in the lab to have specific receptors (CARs) on their surface. These receptors can attach to proteins on leukemia cells. The T cells are then multiplied in the lab and given back into the blood, where they can seek out the leukemia cells and attack them.

3. Stem cell transplant
- Allogeneic Stem cell transplant
- An autologous transplant may be an option for patients who can’t have an allogeneic transplant because they don’t have a matched donor, or for some other reason.

4. Chemo (3 phases)
- Induction, which is short and intensive, usually lasts about a month.
- Consolidation (intensification), which is also intensive, typically lasts for a few months.
- Maintenance (post-consolidation), which is less intensive, typically lasts for about 2 years.

https://www.cancer.org/cancer/acute-lymphocytic-leukemia/treating/typical-treatment.html

5. Radiation Therapy
- if spread to the brain and spinal fluid, or to the testicles 睾丸.
- Radiation to the whole body is often an important part of treatment before a bone marrow or peripheral blood stem cell transplant
- reduce pain in an area of bone invaded by leukemia, if chemotherapy hasn’t helped.
