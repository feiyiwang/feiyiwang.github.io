# <font color='green'>Kernel</font>

Computing the inner product between two feature vectors can be cheap even if the vectors are very high dimensional.
This type of dot function of two feature vectors/examples is a kernel function.

### Example 1

Assume we map $x$  and  $x'∈ℝ^2$  to feature vectors  𝜙(𝑥)  and  𝜙(𝑥′)  given by <br>
$\phi(x)=[{\color{red} {x_1,x_2}},{\color{blue} {{x_1}^2,\sqrt{2}x_1x_2,{x_2}^2}}]$<br>
$\phi(x')=[{\color{red} {x_1',x_2'}},{\color{blue} {{x_1'}^2,\sqrt{2}x_1'x_2',{x_2'}^2}}]$<br>
$\therefore$ &nbsp;&nbsp;&nbsp; a kernel &nbsp;&nbsp;&nbsp;   $K\left \langle x,x' \right \rangle=\phi(x) \cdot \phi(x')={\color{red} {(x\cdot x')}}+{\color{blue} {(x\cdot x')^2}}$<br>

### Example 2

If&nbsp;&nbsp;&nbsp;  $K\left \langle x,x' \right \rangle=\phi(x)\cdot\phi(x')={\color{red} {x_1x_1'}}+{\color{blue} {x_2x_2'+x_2x_3'+x_3x_2'+x_3x_3'}}$<br>
then &nbsp;&nbsp;&nbsp; $\phi(x)=[{\color{red} {x_1}},{\color{blue} {x_2+x_3}}]$ &nbsp;&nbsp;&nbsp; $\phi(x')=[{\color{red} {x_1'}},{\color{blue} {x_2'+x_3'}}]$

# <font color='green'>Kernel Perceptron Algorithm</font>

<b>Perceptron</b> $\left ( \left \{ (x^{(i)},y^{(i)}),i=1,...,n \right \} ,T\right ):$ <br>
  initialize  $\theta=0, \theta_0=0$  (vector);<br>
    for  $𝑡=1,...,𝑇$ ,<br>
      for  $𝑖=1,...,𝑛$ ,<br>
        if  $y^{(i)}(\theta \cdot x^{(i)}+\theta_0)\leqslant 0$ ,<Br>
        then update  $\theta=\theta+y^{(i)}\cdot x^{(i)}, \theta_0=\theta_0+y^{(i)} $

<b>Now</b> $\because \theta = \sum_{j=1}^{n} a_j y^{(j)} \phi(x^{(j)}), \therefore $ we can substitute <br>
    initialize  $\theta=0, \theta_0=0$  (vector);<br>
  with <br>
  initialize  $a_1=...=a_n=0$ ;<br>

<b>Kernel Perceptron</b> $\left ( \left \{ (x^{(i)},y^{(i)}),i=1,...,n \right \} ,T\right ):$ <br>
  initialize  $a_1,...,a_n $ to some value ;<br>
    for  $𝑡=1,...,𝑇$ ,<br>
      for  $𝑖=1,...,𝑛$ ,<br>
        if  $y^{(i)}(\sum_{j=1}^{n} a_j y^{(j)} \phi(x^{(j)})\phi(x^{(i)})+\theta_0)\leqslant 0$ ,<Br>
        then update  $a_i=a_i+i $


Here $K\left \langle x^{(j)},x^{(i)} \right \rangle=\phi(x^{(j)}) \cdot \phi(x^{(i)})$

### Composition rules:
1. $K\left \langle x,x' \right \rangle = 1$ is a kernel function. <Br>$\therefore \phi(x)=1$
2. Let $f:\mathbb{R}^d \rightarrow \mathbb{R}$ , where $f(x)$ is a scalr, and  $K\left \langle x,x' \right \rangle$  is a kernel. Then so is
$\widetilde{K}\left \langle x,x' \right \rangle=f(x)K\left \langle x,x' \right \rangle f(x')$.<br> $\therefore \phi(x)=1 \cdot f(x)$ &nbsp;&nbsp;&nbsp; or say &nbsp;&nbsp;&nbsp; $ \varphi(x) = \phi(x) \cdot f(x)$
3. If $K_1\left \langle x,x' \right \rangle$ and $K_2\left \langle x,x' \right \rangle$ are kernels, then
$K\left \langle x,x' \right \rangle = K_1\left \langle x,x' \right \rangle + K_2\left \langle x,x' \right \rangle$ is a kernel
4. If $K_1\left \langle x,x' \right \rangle$ and $K_2\left \langle x,x' \right \rangle$ are kernels, then $K\left \langle x,x' \right \rangle = K_1\left \langle x,x' \right \rangle K_2\left \langle x,x' \right \rangle$ is a kernel

### Example 1

If  $f:\mathbb{R}^d \rightarrow \mathbb{R}$ and  $K\left \langle x,x' \right \rangle$  is a kernel, &nbsp;&nbsp;&nbsp; so is $\widetilde{K}\left \langle x,x' \right \rangle=f(x)K\left \langle x,x' \right \rangle f(x')$. <br>
 
If $K\left \langle x,x' \right \rangle=\phi(x) \cdot \phi(x')$ and $\widetilde{K}\left \langle x,x' \right \rangle=\varphi(x) \cdot \varphi(x')$ &nbsp;&nbsp;&nbsp; then $\varphi(x)=f(x)\phi(x)$

### Example 2

1. 1 is a kernel for &nbsp;&nbsp;&nbsp; $\phi(x)=1$
2. $x\cdot x'$ is a kernel for &nbsp;&nbsp;&nbsp; $\phi(x)=x$
3. $1+x \cdot x'$ is a kernel since the sum of kernels are still kernels for &nbsp;&nbsp;&nbsp; $\phi(x)=[1,x]^T$
4. $(1+x \cdot x')^2$ is a kernel since the product of kernels are kernels. (In this case, factoring the kernel as dot products are more involved, and the composition rule saves this work.)
5. $exp(x+x')$ is a kernel for &nbsp;&nbsp;&nbsp; $\phi(x)=exp(x)$
6. $min(x,x') $ for $  x, x' \in \mathbb{Z}$ is <b>NOT</b> a kernel. <br> For example,  min(−1,−1)=−1<0  and hence cannot be written as a dot product and is not a valid kernel.
